```python
import pandas as pd
df = pd.read_excel('./data/gallery_list.xlsx', index_col='id', keep_default_na=False)
display(df)
```

```python
from collections import Counter
stack = set()
counter = Counter()
for idx, row in df.iterrows():
    if not row['event']:
        counter[(row['title'], row['date'])]+=1
        
for entry, count in counter.most_common():
    title, date = entry
    if count==2:
        print(count, title, date)
```
