# Scrape ikon

```python
"""read exhibition list"""
import pandas as pd
df = pd.read_excel("./data/exhibition_list_clean_v000.xlsx", index_col='id')
display(df)
```

```python
"""with each ikonid scrape ikon for exhibitions and save details html to disk"""
import mechanicalsoup
import os
import time
browser = mechanicalsoup.StatefulBrowser()
progress=0
for idx, row in df.iterrows():
    ikonid = row['ikonid']
    filepath = "./tmp/exhibition_details/exhibition{}.html".format(ikonid)
    print("{}/{} - {}".format(progress, len(df), filepath))
    progress+=1
    
    # pass existing files
    if os.path.exists(filepath):
        print("...exist")
        continue
    
    # wait 10sec if not responding
    url = "https://ikon.hu/cal/{}".format(ikonid)
    while True:
        response = browser.open(url)
        if response.status_code == 200:
            break
        else:
            print("...wait 10 seconds...")
            time.sleep(3)
            
    # save html to disk
    soup = browser.get_current_page()
    with open(filepath, 'w', encoding='utf8') as html_file:
        html_file.write(str(soup))
    print("...saved")
    
```

# Parse ikon details

```python
import os
from bs4 import BeautifulSoup
from IPython.display import display, IFrame, HTML

data = []
dirpath = "./tmp/exhibition_details"
filepaths = os.listdir(dirpath)
progress = 0
for filename in filepaths:
    print("{}/{}".format(progress, len(filepaths)), end="\r")
    progress+=1
    ikonid = filename[len("exhibition"):-len(".html")]
    url = "https://ikon.hu/cal/{}".format(ikonid)
    filepath = dirpath+"/"+filename
    
    row = {'ikonid': ikonid}
    with open(filepath, encoding="utf-8") as f:
        soup = BeautifulSoup(f.read())
        soup = soup.find(class_="oneEvent")
            
        # Parse details
        detailselem = soup.find_all("div", class_="row")[1].find('div', class_="rightcol")
        for elemrow in detailselem.find_all("p", attrs={'class': None}):                
            if "|" not in elemrow.text:
                continue
                
            key, value = elemrow.text.split("|", 1)
            key, value = key.strip(), value.strip()
            row[key] = value
            
        data.append(row)

import pandas as pd
df = pd.DataFrame(data)
df = df.set_index('ikonid')
display(df)
```
